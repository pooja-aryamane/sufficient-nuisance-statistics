## GROUPWISE BEST ACCURACIES - OF THE BEST MODEL

### 1. true group: worst val accuracy = 75% (on mimic 0?) 

```python
def get_truegroupwise_acc(df, path):
    by_group=df.groupby('true_group').apply(lambda group: (group['predictions'] == group['gt']).sum() / len(group))
    return by_group
truegroup_valpath = 'grouptrue_group_idx+weightdecay_0.01+pretrained_True+batchsize_256+lr_0.0001+transform_True_val_preds_maxacc.csv'
val_true_preds = pd.read_csv(f'stage2-outdir/{truegroup_valpath}').iloc[:,1:]
get_truegroupwise_acc(val_true_preds, 'val_path') #75% is the worst group accuracy 
```

### 2. sns groups:

```python
def get_worstgroup_acc(df, path):
    by_group=df.groupby('true_group').apply(lambda group: (group['predictions'] == group['gt']).sum() / len(group))
    worst_group = by_group.idxmin()
    return worst_group,by_group[worst_group]
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
def get_acc(df):
    acc = (df['snsy4_pred_label'] == df['Cardiomegaly']).sum() / df.shape[0]
    return acc
```
