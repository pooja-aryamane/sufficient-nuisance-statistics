### 2. QUANTILE SPLITS

#### load new stage 1 probs

```python
#NEW STAGE 1 : get probs: run with new stage 1 model 
def sigmoid(x):                                        
    return 1 / (1 + np.exp(-x))
split1=pd.read_csv('stage1-outdir/split_0+decay_0.01+pretrained_True+batchsize_256+lr_0.001+transform_True_opposite_split_predictions.csv').iloc[:,1:]
split0=pd.read_csv('stage1-outdir/split_1+decay_0.01+pretrained_True+batchsize_256+lr_0.001+transform_True_opposite_split_predictions.csv').iloc[:,1:]
split0['predprob_new'] = sigmoid(split0.max_accuracy_model_predictions)
split1['predprob_new'] = sigmoid(split1.max_accuracy_model_predictions)
combined_splits = pd.concat([split0,split1])
y = combined_splits.Cardiomegaly.values
combined_splits['observed_prob'] = y * combined_splits.predprob_new + (1-y)* (1-combined_splits.predprob_new)
df = combined_splits.copy()

#create groups: 
for n in [2,4,8,16,32]:
    n_groups = n
    group_label = 'sns_group_'+str(n)
    
    df[group_label]=pd.qcut(df.observed_prob,n_groups,labels=False)
    
    x_group = df[group_label].values.reshape(-1,1)
    y_group = df['true_group_idx'].values
    
    clf = DecisionTreeClassifier(random_state=0, max_depth=5, class_weight='balanced')
    clf = clf.fit(x_group, y_group)
    print(f'sns {n} groups classification = ')
    print(balanced_accuracy_score(clf.predict(x_group), y_group))

#check group wise accuracies (majority group classification) 
for x in np.unique(x_group.reshape(-1)):
    print(x, y_group[x_group.reshape(-1)==x].mean())
```

```python
chex = pd.read_csv('resized_data/chexpert/full_data_chexpert_full.csv').iloc[:,1:]
mimic = pd.read_csv('resized_data/mimic/full_data_mimic_full.csv').iloc[:,1:]
df=pd.concat([chex,mimic])
p = 0.1
def generate_noisy_group(group_idx):
    if np.random.rand() < p:
        return np.random.choice([idx for idx in range(4) if idx != group_idx])
    else:
        return group_idx

df['noisy_true_groups'] = df['true_group_idx'].apply(generate_noisy_group)
print((df['noisy_true_groups']==df['true_group_idx']).sum()/df.shape[0])
```
