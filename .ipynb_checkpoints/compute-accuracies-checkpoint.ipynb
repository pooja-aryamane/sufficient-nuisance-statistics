```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def get_acc(df):
    df['pred_prob'] = sigmoid(df.predictions)
    df['pred_label'] = [1 if p>0.5 else 0 for p in df.pred_prob]
    acc = (df['pred_label'] == df['Cardiomegaly']).sum() / df.shape[0]
    return acc
```

#### WORST GROUP ACC

```python
acc_dfs=[]
experiments=[]
for i,path in enumerate(os.listdir('balanced-erm-new/fullpreds')):
    if path.startswith('.'):
        pass
    else:
        wg={}
        pred_df = pd.read_csv('balanced-erm-new/fullpreds/'+path, low_memory=False)
        t_acc = pred_df[pred_df.split=='test'].groupby('true_group_idx').apply(get_acc)
        v_acc = pred_df[pred_df.split=='val'].groupby('true_group_idx').apply(get_acc)
        t_wg,t_wa = t_acc.idxmin(), t_acc.min()
        v_wg,v_wa = v_acc.idxmin(), v_acc.min()
        wg['t_wg'] = t_wg
        wg['t_wa'] = t_wa
        wg['v_wg'] = v_wg
        wg['v_wa'] = v_wa
        wg_acc_df = pd.DataFrame(wg,index=[i])
        acc_dfs.append(wg_acc_df)
        experiments.append(path)
acc_df=pd.concat(acc_dfs)
acc_df['experiment'] = experiments
acc_df=pd.concat([acc_df.experiment.str.split('+',expand=True),acc_df],axis=1)
acc_df=acc_df.drop(['experiment'],axis=1)
```

```python
acc_df=acc_df.rename(columns={0:'ngroups',1:'wd',2:'pretrained',3:'batch_size',4:'lr',5:'experiment'})
```

```python
acc_df['weight_decay'] = acc_df.wd.str.split('_').str[-1]
acc_df['learning_rate']= acc_df.lr.str.split('_').str[-1]
```
